```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Importing the dataset

df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/Tinder%20Millennial%20Match%20Rate%20Dataset.csv')
df.head()
```

```python
# Convert to percentage 

df['MatchPercent'] = df['Percentage']*100
df.head()
```

```python
# Finding all the unique values from the MatchPercent column

df['MatchPercent'].unique()
```

```python
# Categorizing percentages

def changePercent(num):
  if num >= 0 and num <=20.9:
    return "0-20%"

  elif num >= 21 and num <=40:
    return "21-40%"
  
  elif num >= 41 and num <=60:
    return "41-60%"
  
  elif num >= 61 and num <=80:
    return "61-80%"
  
  elif num >= 81:
    return "80% and up"

df['SuccessRange'] = df['MatchPercent'].apply(lambda num:changePercent(num))
df.head()
```

```python
# Presenting the success rate of Tinder users

sns.countplot(x=df['SuccessRange'])
```

```python
# Responses of Tinder users on whether it ended up in a relationship.

sns.countplot(x=df['It became a Relationship'])
```

```python
# Change MatchCount formatting

def findMatchCount(num):
  if num>=0 and num <=5:
    return "0-5 matches"
  
  elif num>=6 and num<=20:
    return "6-20 matches"
  
  elif num>=21 and num<=50:
    return "21-50 matches"
  
  elif num>50:
    return "50+ matches"

df['HowManyMatches'] = df['Count'].apply(lambda num:findMatchCount(num))
df.head()
```

```python
# Match count of Tinder users

sns.countplot(x=df['HowManyMatches'])
```
