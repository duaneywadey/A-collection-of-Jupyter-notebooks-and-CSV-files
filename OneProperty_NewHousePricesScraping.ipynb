```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = "https://onepropertee.com/properties?propertyType=Condominiums-and-Apartments"
```

```python
# Get request

response = requests.get(website)
```

```python
# Get status code

response.status_code
```

```python
# Store the website's HTML file to a variable

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
results = soup.find_all('div', {'class':'listing'})
print(len(results))
```

```python
house_desc = []
house_loc = []
house_price = []

for i in range(1,251):
  website = 'https://onepropertee.com/condo-for-sale-philippines/page/' + str(i)

  response = requests.get(website)

  soup = BeautifulSoup(response.content, 'html.parser')

  results = soup.find_all('div', {'class':'listing'})

  for result in results:
    try:
      a = result.find('div', {'class': 'listing-field listing-requiredfields'}).get_text()
      b = result.find('div', {'class': 'listing-field listing-location'}).get_text()
      c = result.find('div', {'class': 'listing-field listing-price d-flex justify-content-between'}).get_text()
      house_desc.append(a)
      house_loc.append(b)
      house_price.append(c)
    except:
      print('N/A')
```

```python
# Store it inside Pandas dataframe

product_overview = pd.DataFrame({'HouseDescription': house_desc, 'Address': house_loc, 'Cost': house_price})
product_overview.head(20)
```

```python
# Save as CSV File

product_overview.to_csv('newCondoPricesPH.csv', index=False)
```
