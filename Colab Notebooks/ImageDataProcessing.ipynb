```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading the image through matplotlib.image module

img = mpimg.imread('/content/322868_1100-800x825.jpg')
```

```python
# get the data type of img

type(img)
```

```python
# shape of the img

print(img.shape)
```

```python
# display the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

```python
import cv2
```

```python
img = cv2.imread('/content/322868_1100-800x825.jpg')
```

```python
type(img)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
cv2_imshow(grayscale_image)
```
