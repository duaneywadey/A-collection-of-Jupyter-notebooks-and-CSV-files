```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/report.csv')
df.head()
```

```python
# Group by report year, display homicide mean

plt.figure(figsize=(18,5))
df.groupby('report_year')['homicides'].mean()
df.groupby('report_year')['homicides'].mean().plot()
```

```python
# Robbery cases over the years

plt.figure(figsize=(20,5))
df.groupby('report_year')['rapes'].mean()
df.groupby('report_year')['rapes'].mean().plot()
```

```python
# Robbery cases all over the years

plt.figure(figsize=(20,5))
df.groupby('report_year')['robberies'].mean()
df.groupby('report_year')['robberies'].mean().plot()
```

```python
# Top 5 years with the most number of homicides

plt.figure(figsize=(20,5))
df.groupby('report_year')['homicides'].mean().nlargest(5).plot.bar()
```

```python
# Top 5 years with the most number of rapes

plt.figure(figsize=(20,5))
df.groupby('report_year')['rapes'].mean().nlargest(5).plot.bar()
```

```python
# Top 10 states with most number of rape cases

plt.figure(figsize=(20,5))
df.groupby('agency_jurisdiction')['rapes'].mean().nlargest(10).plot.bar()
```

```python
# Top 10 states with most number of robbery cases

plt.figure(figsize=(20,5))
df.groupby('agency_jurisdiction')['robberies'].mean().nlargest(10).plot.bar()
```

```python
# Top 10 states with least cases of rape

plt.figure(figsize=(20,5))
df.groupby('agency_jurisdiction')['rapes'].mean().nsmallest(10).plot.bar()
```

```python
# Top 10 states with least cases of robbery

plt.figure(figsize=(20,5))
df.groupby('agency_jurisdiction')['robberies'].mean().nsmallest(10).plot.bar()
```

```python
# Top 10 states with least cases of homicide

plt.figure(figsize=(20,5))
df.groupby('agency_jurisdiction')['homicides'].mean().nsmallest(10).plot.bar()
```
