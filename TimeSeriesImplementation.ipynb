```python
# Add new features to our dataset

features = df
features['year'] = df['date'].dt.year
features['month'] = df['date'].dt.month
features['day'] = df['date'].dt.day
features.head(5)
```

```python
features['lag1'] = df['births'].shift(1)
features['lag2'] = df['births'].shift(365)
features.head()
```

```python
# Get the mean from the preceding observation
features['Roll_mean'] = df['births'].rolling(2).mean()
features.head()
```

```python
features.head()
```

```python
# Get the maximum from the preceding three observations
features['Roll_max'] = df['births'].rolling(3).max()
```

```python
# Get the maximum value (so far)
features['Expand_max'] = df['births'].expanding().max()
```
