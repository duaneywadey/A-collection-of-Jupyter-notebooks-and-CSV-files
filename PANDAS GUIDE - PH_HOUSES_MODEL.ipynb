```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/CleanHousesData.csv')
```

```python
def AddressToRegion(address):
  RegionOne = ['Ilocos Norte', 'Ilocos Sur', 'La Union', 'Pangasinan']
  r1 = [item.lower() for item in RegionOne]

  RegionTwo = ['Cagayan', 'Isabela', 'Nueva Vizcaya', 'Quirino', 'Batanes']
  r2 = [item.lower() for item in RegionTwo]

  RegionThree = ['Aurora', 'Bataan', 'Bulacan', 'Nueva Ecija', 'Pampanga', 'Tarlac', 'Zambales']
  r3 = [item.lower() for item in RegionThree]

  RegionFourA = ['Cavite', 'Laguna', 'Batangas', 'Rizal', 'Quezon']
  r4a = [item.lower() for item in RegionFourA]

  RegionFourB = [ 'Occidental Mindoro', 'Oriental Mindoro', 'Marinduque', 'Romblon', 'Palawan']
  r4b = [item.lower() for item in RegionFourB]

  RegionFive = ['Albay', 'Camarines Norte', 'Camarines Sur', 'Sorsogon', 'Catanduanes', 'Masbate']
  r5 = [item.lower() for item in RegionFive]

  RegionSix = ['Aklan', 'Antique', 'Negros Occidental', 'Capiz', 'Guimaras', 'Iloilo']
  r6 = [item.lower() for item in RegionSix]

  RegionSeven = ['Bohol', 'Cebu', 'Negros Oriental', 'Siquijor']
  r7= [item.lower() for item in RegionSeven]
  
  RegionEight = ['Biliran', 'Eastern Samar', 'Leyte', 'Northern Samar', 'Samar', 'Southern Leyte']
  r8 = [item.lower() for item in RegionEight]

  RegionNine = ['Zamboanga del Norte', 'Zamboanga Sibugay', 'Zamboanga del Sur']
  r9 = [item.lower() for item in RegionNine]

  RegionTen = ['Bukidnon', 'Camiguin', 'Lanao del Norte', 'Misamis Occidental', 'Misamis Oriental']
  r10 = [item.lower() for item in RegionTen]

  RegionEleven = ['Davao de Oro', 'Davao del Norte', 'Davao del Sur', 'Davao Oriental', 'Davao Occidental']
  r11 = [item.lower() for item in RegionEleven]

  RegionTwelve = ['Cotabato', 'South Cotabato', 'Sultan Kudarat', 'Sarangani']
  r12 = [item.lower() for item in RegionTwelve]

  RegionThirteen = ['Agusan del Norte', 'Agusan del Sur', 'Dinagat Isands', 'Surigao del Sur', 'Surigao del Norte']
  r13 = [item.lower() for item in RegionThirteen]
  
  CAR = ['Abra', 'Apayao', 'Ifugao', 'Benguet', 'Mountain Province','Kalinga']
  rCAR = [item.lower() for item in CAR]

  BARMM = ['Basilan', 'Lanao del Sur', 'Maguindanao', 'Sulu', 'Tawi-tawi']
  rBARMM = [item.lower() for item in BARMM]

  mnl = 'Manila'

  if any(substring in address.lower() for substring in r1):
    reg1 =  'I'
    return reg1

  elif any(substring in address.lower() for substring in r2):
    reg2 = 'II'
    return reg2
  
  elif any(substring in address.lower() for substring in r3):
    reg3 = 'III'
    return reg3
  
  elif any(substring in address.lower() for substring in r4a):
    reg4a = 'IV A'
    return reg4a
  
  elif any(substring in address.lower() for substring in r4b):
    reg4b = 'IV B'
    return reg4b
  
  elif any(substring in address.lower() for substring in r5):
    reg5 = 'V'
    return reg5
  
  elif any(substring in address.lower() for substring in r6):
    reg6 = 'VI'
    return reg6
  
  elif any(substring in address.lower() for substring in r7):
    reg7 = 'VII'
    return reg7
    
  elif any(substring in address.lower() for substring in r8):
    reg8 = 'VIII'
    return reg8
  
  elif any(substring in address.lower() for substring in r9):
    reg9 = 'IX'
    return reg9
  
  elif any(substring in address.lower() for substring in r10):
    reg10 = 'X'
    return reg10
  
  elif any(substring in address.lower() for substring in r11):
    reg11 = 'XI'
    return reg11
  
  elif any(substring in address.lower() for substring in r12):
    reg12 = 'XII'
    return reg12
  
  elif any(substring in address.lower() for substring in r13):
    reg13 = 'XIII'
    return reg13
  
  elif any(substring in address.lower() for substring in rCAR):
    car = 'CAR'
    return car
  
  elif any(substring in address.lower() for substring in rBARMM):
    barmm = 'BARMM'
    return barmm
  
  elif mnl.lower() in address.lower():
    mnla = 'NCR'
    return mnla

```

```python
df['Region'] = df['Address'].apply(lambda text: AddressToRegion(text))
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df['Region'].value_counts()
```

```python
df[df.isnull().any(axis=1)]
```

```python
plt.figure(figsize=(20,5))
df.groupby('Region')['Cost'].mean().nlargest(10).plot.bar()
```

```python
dummies = pd.get_dummies(df.Region)
```

```python
merged = pd.concat([df, dummies],axis='columns')
merged.drop(columns=['Region', 'Address'], inplace=True)
merged.head()
```

```python
final = merged.drop(['BARMM'], axis='columns')
final
```

```python
X = final.drop('Cost', axis='columns')
X
```

```python
Y = final.Cost
Y
```

```python
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
# Split our training and test data, random_state - to split the data

X = X.to_numpy()
Y = Y.to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, random_state=2)
```

```python
# Let's now train our model
lin_reg_model = XGBRegressor()
lin_reg_model.fit(X_train,Y_train)
```

```python
# Evaluate our model, predicts the price values depending on other data given
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
from sklearn import metrics

# R squared error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared error: ", error_score)
```

```python
# Visualize the actual prices and predicted prices

plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual Prices")
```

```python
# Evaluate our model, predicts the price values depending on other data given
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error: ", error_score)
```

```python
# Visualize the actual prices and predicted prices

plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual Prices")
```
