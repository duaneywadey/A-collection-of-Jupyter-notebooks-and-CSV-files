```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Store website in variable 

website = 'https://www.abenson.com/mobile/smartphone/page/2.html'
```

```python
# Get request

response = requests.get(website)
```

```python
# Get status code

response.status_code
```

```python
# Store the website's HTML file to a variable

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
results = soup.find_all('li', {'class':'item product product-item item-custom-product'})
print(len(results))
```

```python
# Product name

product_name = []
product_specs = []
product_costs = []

for i in range(2,24):
  website = 'https://www.abenson.com/mobile/smartphone/page/' + str(i) + '.html'

  response = requests.get(website)

  soup = BeautifulSoup(response.content, 'html.parser')

  results = soup.find_all('li', {'class':'item product product-item item-custom-product'})
  
  for result in results:
    try:
      a = result.find('a', {'class': 'product-item-link'}).get_text()
      b = result.find('div', {'class': 'product description product-item-description'}).get_text()
      c = result.find('span', {'class': 'price'}).get_text()
      product_name.append(a)
      product_specs.append(b)
      product_costs.append(c)

    except:
      print('N/A')
    
```

```python
# Store it inside Pandas dataframe

product_overview = pd.DataFrame({'ProductDescription': product_name, 'ProductSpecs': product_specs , 'Prices': product_costs})
product_overview.head(20)
```

```python
# Save as CSV File

product_overview.to_csv('phone_ph_prices.csv', index=False)
```
