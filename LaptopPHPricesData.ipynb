```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Store website in variable 

website = 'https://asianic.com.ph/product_list/notebooks'
```

```python
# Get request

response = requests.get(website)
```

```python
# Get status code

response.status_code
```

```python
 # Store the website's HTML file to a variable

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
# Find class containing the data to be scraped

results = soup.find_all('div', {'class':'prod_container'})
len(results)
```

```python

# Product name

for i in range(235):
  print(results[i].find('div', {'class': 'prod_col2'}).get_text())
```

```python

# Laptop prices

for i in range(235):
  print(results[i].find('div', {'class': 'prod_col3'}).get_text())
```

```python
# Add product names inside a list

product_name = []

for result in results:
  product_name.append(result.find('div', {'class': 'prod_col2'}).get_text())

product_name
```

```python

# Laptop prices

product_prices = []

for result in results:
  product_prices.append(result.find('div', {'class': 'prod_col3'}).get_text())

product_prices
```

```python
# Store it inside Pandas dataframe

product_overview = pd.DataFrame({'ProductDescription': product_name, 'Prices': product_prices})
product_overview.head()
```

```python
# Save as CSV File

product_overview.to_csv('laptop_ph_prices.csv', index=False)
```
