```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/phone_ph_prices.csv')
df.head()
```

```python
df['Prices'] = df['Prices'].str.replace('₱', '')
df['Prices'] = df['Prices'].str.replace(',', '')
df['Prices'] = df['Prices'].astype('float64')

df.head()
```

```python
# Getting rid of all '\n'

df['ProductDescription'] = df['ProductDescription'].str.replace('\n', '')
df['ProductSpecs'] = df['ProductSpecs'].str.replace('\n', '')
df.head()
```

```python
df['Brand'] = df['ProductDescription'].apply(lambda text: text.split()[0])
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Brand'], y=df['Prices'])
```

```python
sns.countplot(df['Brand'])
```

```python
df['ScreenSizes'] = df['ProductSpecs'].apply(lambda text: text.split()[0])
df.head()
```

```python
df.sample(10)
```

```python
df['ScreenSizes'].unique()
```

```python
df = df.drop(df[df.Brand == 'Apple'].index)
df = df.drop(df[df.ScreenSizes == 'Digital'].index)
df = df.drop(df[df.ScreenSizes == 'Cordless'].index)
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(df['Brand'])
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Brand'], y=df['Prices'])
```

```python
df.sample(10)
```

```python
df['ScreenSizes'] = df['ScreenSizes'].str.replace('-inch', '')
df['ScreenSizes'] = df['ScreenSizes'].str.replace(',', '')
df['ScreenSizes'] = df['ScreenSizes'].str.replace('es', '')
df['ScreenSizes'] = df['ScreenSizes'].astype('float64')
```

```python
df.sample(5)
```

```python
sns.distplot(df['ScreenSizes'])
```

```python
sns.distplot(df['Prices'])
```

```python
df['ZScore'] = (df.Prices - df.Prices.mean()) / df.Prices.std()
sns.distplot(df['ZScore'])
```

```python
# Displaying the outliers 

values = df[(df.ZScore>3) | (df.ZScore<-3)]
values
```

```python
def addStorageColumn(specs):
  storage1 = '32GB'
  storage2 = '64GB'
  storage3 = '128GB'
  storage4 = '256GB'
  storage5 = '512GB'

  storageOthers = 'Others'

  if storage1.lower() in specs.lower():
    return storage1
  
  elif storage2.lower() in specs.lower():
    return storage2
  
  elif storage3.lower() in specs.lower():
    return storage3
  
  elif storage4.lower() in specs.lower():
    return storage4

  elif storage5.lower() in specs.lower():
    return storage5
  
  else:
    return storageOthers
    

# Apply the changes to our dataframe

df['StorageCapacity'] = df['ProductSpecs'].apply(lambda text:addStorageColumn(text))
df.sample(10)
```

```python
df = df.drop(columns=['ZScore'])
df.sample(5)
```

```python
df['ScreenSizes'].sort_values().unique()
```

```python
def changeScreenSizeFormat(num):
  if num >=6.0 and num <=6.5:
    return "6.0-6.5 inches"

  elif num >=6.51 and num <=6.99:
    return "6.5-6.99 inches"
  
  elif num >=7.0:
    return "7.0 inches and up"

# 'values' represents all values inside ScreenSize column
df['ScreenSize'] = df['ScreenSizes'].apply(lambda values: changeScreenSizeFormat(values))
df.head()
```

```python
df['Brand'] = df['Brand'].str.replace('Oppo', 'OPPO')
plt.figure(figsize=(15,8))
sns.countplot(df['Brand'])
```

```python
values = df[(df.Brand == 'Nokia') | (df.Brand == 'Motorola') | (df.Brand == 'Lenovo')]
values
```

```python
df = df.drop(df[df.Brand == 'Nokia'].index)
df = df.drop(df[df.Brand == 'Motorola'].index)
df = df.drop(df[df.Brand == 'Lenovo'].index)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(df['Brand'])
```

```python
df.sample(5)
```

```python
df['BatteryCap'] = df['ProductSpecs'].apply(lambda text: text.split()[-1:])
df['BatteryCap'] = df['BatteryCap'].str.get(0)
df['BatteryCap'] = df['BatteryCap'].str.replace('mAh', '')
df['BatteryCap'] = df['BatteryCap'].str.replace('mAH', '')
df = df.drop(df[df.BatteryCap == '128GB'].index)
df = df.drop(df[df.BatteryCap == ''].index)
df['BatteryCap'] = df['BatteryCap'].astype('int32')
df.head()
```

```python
sns.scatterplot(x=df['BatteryCap'], y=df['Prices'])
```

```python
df['BatteryCap'].sort_values().unique()
```

```python
def changeBatteryCapFormat(batteryCap):

  if batteryCap >= 3700 and batteryCap <= 4200:
    return "3700mAh-4200mAh"
  
  elif batteryCap >= 4201 and batteryCap <= 4999:
    return "4201mAh-4999mAh"
  
  elif batteryCap >= 5000:
    return "5000+mAh"

df['BatteryCap'] = df['BatteryCap'].apply(lambda vals: changeBatteryCapFormat(vals))
```

```python
df = df[['Brand', 'StorageCapacity', 'ScreenSize', 'BatteryCap', 'Prices']]
df.sample(10)
```

```python
df.to_csv('CleanedMobilePricesData.csv')
```
