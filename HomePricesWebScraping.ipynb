```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = "https://www.myproperty.ph/house/buy/"
```

```python
# Get request

response = requests.get(website)
```

```python
# Get status code

response.status_code
```

```python
# Store the website's HTML file to a variable

soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
results = soup.find_all('div', {'class':'row ListingCell-row ListingCell-agent-redesign'})
print(len(results))
```

```python
house_desc = []
house_loc = []
house_bedrooms = []
house_floor_area = []
house_land_size = []

for i in range(1,101):
  website = 'https://www.myproperty.ph/house/buy/?page=' + str(i)

  response = requests.get(website)

  soup = BeautifulSoup(response.content, 'html.parser')

  results = soup.find_all('div', {'class':'row ListingCell-row ListingCell-agent-redesign'})

  for result in results:
    try:
      a = result.find('h2', {'class': 'ListingCell-KeyInfo-title'}).get_text()
      b = result.find('span', {'class': 'ListingCell-KeyInfo-address-text'}).get_text()
      house_desc.append(a)
      house_loc.append(b)
      
    except:
      print('N/A')


```

```python
# Store it inside Pandas dataframe

product_overview = pd.DataFrame({'HouseDescription': house_desc, 'Address': house_loc})
product_overview.head(20)
```

```python
# Save as CSV File

product_overview.to_csv('home_prices_ph.csv', index=False)
```
