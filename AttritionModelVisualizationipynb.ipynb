```python
import numpy as np
import pandas as pd
pd.set_option('max_columns', None)

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/attrition_train.csv')
df.head()
```

```python
catList = df.select_dtypes(include=['object']).columns.tolist()
for c in catList:
  fig, ax = plt.subplots(figsize=(20,5))
  sns.set_theme()
  sns.countplot(df[c])
```

```python
# Historgram representation of the numeric values

numList = df.select_dtypes(include=['int64']).columns.tolist()

for n in numList:
  sns.set_theme()
  g = sns.displot(x=df[n], label=[n], kde=True, height=20,aspect=20)
  g.fig.set_size_inches(20,5)

  
```

```python
# Checking if values are normally distributed

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

numList = df.select_dtypes(include=['int64']).columns.tolist()

for n in numList:
  plt.figure(figsize=(20,5))
  test = df[n].to_numpy()
  h = test
  h.sort()
  hmean = np.mean(h)
  hstd = np.std(h)
  pdf = stats.norm.pdf(h, hmean, hstd)
  plt.xlabel(n)
  plt.plot(h, pdf) # including h here is crucial

```
