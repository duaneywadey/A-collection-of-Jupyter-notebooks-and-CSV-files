```python
df['Region'] = df['Address'].apply(lambda text: AddressToRegion(text))
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df['Region'].value_counts()
```

```python
df[df.isnull().any(axis=1)]
```

```python
plt.figure(figsize=(20,5))
df.groupby('Region')['Cost'].mean().nlargest(10).plot.bar()
```

```python
dummies = pd.get_dummies(df.Region)
```

```python
merged = pd.concat([df, dummies],axis='columns')
merged.drop(columns=['Region', 'Address'], inplace=True)
merged.head()
```

```python
final = merged.drop(['Other City/Province'], axis='columns')
final
```

```python
X = final.drop('Cost', axis='columns')
X
```

```python
Y = final.Cost
Y
```

```python
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
# Split our training and test data, random_state - to split the data

X = X.to_numpy()
Y = Y.to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, random_state=2)
```

```python
# Let's now train our model
lin_reg_model = XGBRegressor()
lin_reg_model.fit(X_train,Y_train)
```

```python
# Evaluate our model, predicts the price values depending on other data given
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
from sklearn import metrics

# R squared error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared error: ", error_score)
```

```python
# Visualize the actual prices and predicted prices

plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual Prices")
```

```python
# Evaluate our model, predicts the price values depending on other data given
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error: ", error_score)
```

```python
# Visualize the actual prices and predicted prices

plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual Prices")
```
