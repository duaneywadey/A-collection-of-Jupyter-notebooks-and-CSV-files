```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(25,8))
sns.set_style('darkgrid')
sns.countplot(df['Employment'])
```

```python
df['FormalEducation'].unique()
```

```python
df['FormalEducation'] = df['FormalEducation'].replace('Bachelor’s degree (BA, BS, B.Eng., etc.)', 'BS degree')
df['FormalEducation'] = df['FormalEducation'].replace('Master’s degree (MA, MS, M.Eng., MBA, etc.)', 'MS degree')
df['FormalEducation'] = df['FormalEducation'].replace('Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)', 'Secondary school')
df['FormalEducation'] = df['FormalEducation'].replace('I never completed any formal education', 'No formal education')
df['FormalEducation'] = df['FormalEducation'].replace('Other doctoral degree (Ph.D, Ed.D., etc.)', 'Other doctoral degree')
df['FormalEducation'] = df['FormalEducation'].replace('Some college/university study without earning a degree', 'Undergrad')

plt.figure(figsize=(25,8))
sns.set_style('darkgrid')
sns.countplot(df['FormalEducation'])
```

```python

df['SexualOrientation'].unique()
```

```python
def changeSexOrientation(gender):
  a = 'Asexual'
  b = 'Gay'
  c = 'Straight'

  if a.lower() in gender.lower() and c.lower() not in gender.lower():
    return a
  
  elif b.lower() in gender.lower() and c.lower() not in gender.lower():
    return b
  
  elif c.lower() in gender.lower():
    return c


df['SexualOrientation'] = df['SexualOrientation'].astype('str')
df['SexualOrientation'] = df['SexualOrientation'].apply(lambda text: changeSexOrientation(text))
df.sample(10)
```

```python
plt.figure(figsize=(20,8))
sns.set_style('darkgrid')
sns.countplot(df['SexualOrientation'])
```

```python
def changeGender(gender):
  a = 'Male'
  b = 'Female'
  c = 'Others'

  if a.lower() in gender.lower():
    return a
  
  elif b.lower() in gender.lower():
    return b
  
  else:
    return c


df['Gender'] = df['Gender'].astype('str')
df['Gender'] = df['Gender'].apply(lambda text: changeGender(text))
df.sample(10)
```

```python
plt.figure(figsize=(20,8))
sns.set_style('darkgrid')
sns.countplot(df['Gender'])
```
