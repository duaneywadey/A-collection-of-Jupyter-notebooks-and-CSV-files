```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Store website in variable 

website = 'https://www.centrecom.com.au/laptops?pagesize=96'
```

```python
# Get request

response = requests.get(website)
```

```python
# Get status code

response.status_code
```

```python
 # Store the website's HTML file to a variable

 soup = BeautifulSoup(response.content, 'html.parser')
 soup
```

```python
# Find class containing the data to be scraped

results = soup.find_all('div', {'class':'prbox_box prbox_img'})
len(results)
```

```python
# Check the first result 

results[0]
```

```python

# Product name

for i in range(96):
  print(results[i].find('div', {'class': 'prbox_name'}).get_text())
```

```python
# Product prices

for i in range(96):
  print(results[i].find('div', {'class': 'saleprice'}).get_text())
```

```python
# Add product names inside a list

product_name = []

for result in results:
  product_name.append(result.find('div', {'class': 'prbox_name'}).get_text())
```

```python
product_name
```

```python
# Adding all product costs inside a list

product_cost = []

for result in results:
  product_cost.append(result.find('div', {'class': 'saleprice'}).get_text())
```

```python
product_cost
```

```python
# Store it inside Pandas dataframe

product_overview = pd.DataFrame({'Name': product_name, 'Price': product_cost})
product_overview.head()
```

```python
# Save as CSV File

product_overview.to_csv('file_name.csv', index=False)
```
