```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/duaneywadey/A-collection-of-Jupyter-notebooks-and-CSV-files/main/CSV%20Files/aircraftdata.csv')
df.head()
```

```python
plt.figure(figsize=(18,5))
df['MilesMM'].plot()
```

```python
# Zooming in

dataviz = df[(df['Month'] > '1963-01-01') & (df['Month'] < '1964-01-01')]
dataviz
```

```python
plt.figure(figsize=(20,5))
dataviz['MilesMM'].plot()
```

```python
# Trendline

plt.figure(figsize=(18,5))
sns.regplot(x=df.index.values, y=df['MilesMM'])
```
